```python
#TASK5
#Display the car mileages for the following four companies using a pivot table: Hyundai, Mahindra, Renault, and Skoda
#Create 4 separate pivot tables capturing the following details:
#1. Mean of city mileage
#2. Median of city mileage
#3. Mean of highway mileage
#4. Median of highway mileage

companies = ['Hyundai', 'Mahindra', 'Renault', 'Skoda']
filtered_data = df[df['Make'].isin(companies)]

mean_city_mileage = pd.pivot_table(filtered_data, values='City_Mileage_km_litre', index='Make', aggfunc='mean')

median_city_mileage = pd.pivot_table(filtered_data,  values='City_Mileage_km_litre',  index='Make',  aggfunc='median')

mean_highway_mileage = pd.pivot_table(filtered_data,values='Highway_Mileage_km_litre',  index='Make',  aggfunc='mean')

median_highway_mileage = pd.pivot_table(filtered_data,  values='Highway_Mileage_km_litre',  index='Make',  aggfunc='median')

display("mean_city_mileage",mean_city_mileage)
display("median_city_mileage",median_city_mileage)
display("mean_highway_mileage",mean_highway_mileage)
display("median_highway_mileage",median_highway_mileage)
```

```python
#TASK7 #Display the average mileage (city and highway) of all manufacturers using a horizontal bar chart, in ascending order of mileage.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
average_mileage = df.groupby('Make').agg( Average_City_Mileage=('City_Mileage_km_litre', 'mean'),Average_Highway_Mileage=('Highway_Mileage_km_litre', 'mean'))
average_mileage['Average_Total_Mileage'] = (average_mileage['Average_City_Mileage'] + average_mileage['Average_Highway_Mileage']) / 2

# Sort by average total mileage in ascending order
average_mileage_sorted = average_mileage.sort_values('Average_Total_Mileage')

# Plotting a horizontal bar chart
plt.barh(average_mileage_sorted.index, average_mileage_sorted['Average_Total_Mileage'], color='skyblue')
plt.title('Average Mileage of All Manufacturers (City and Highway)')
plt.xlabel('Average Mileage (km/litre)')
plt.ylabel('Manufacturer')
plt.tight_layout() 
plt.show()
```

```python
 #TASK11 #Preprocess the data in the following columns as specified. Kerb_weight: Remove "kg" and ensure the column is numerical. Convert the weight in kilograms to pounds, by multiplying each value in the column with the number 2.2. Gross_Vehicle_Weight: Remove "kg" and ensure the column is numerical, Convert the weight in kilograms to pounds, by multiplying each value in the column. with the number 2.2.
df['Gross_Vehicle_Weight'] = df['Gross_Vehicle_Weight'].replace('Not Available', np.nan)
df['Gross_Vehicle_Weight'] = df['Gross_Vehicle_Weight'].astype(str)
df['Gross_Vehicle_Weight'] = df['Gross_Vehicle_Weight'].str.replace('kg', '', regex=False)

df['Gross_Vehicle_Weight'] = pd.to_numeric(df['Gross_Vehicle_Weight'], errors='coerce')
df['Weight_in_pounds'] = df['Gross_Vehicle_Weight'] * 2.2
df['Weight_in_pounds']
```

```python
df['Kerb_Weight'] = df['Kerb_Weight'].replace('Not Available', np.nan)
df['Kerb_Weight'] = df['Kerb_Weight'].astype(str)
df['Kerb_Weight'] = df['Kerb_Weight'].str.replace('kg', '', regex=False)

df['Kerb_Weight'] = pd.to_numeric(df['Kerb_Weight'], errors='coerce')
df['Weight_in_pounds'] = df['Kerb_Weight'] * 2.2
df['Weight_in_pounds']
```

```python
#TASK13 #Identify at least three features that affect the mileages (using correlation analysis) of cars from the following manufacturers: Renault, Toyota, and Honda

#Represent them visually using appropriate types of charts/graphs in Python.

#Illustrate with justification whether these factors affect the mileage of a car positively or negatively
df_filtered = df[df['Make'].isin(['Renault', 'Toyota', 'Honda'])]
mileage_df = df_filtered[['City_Mileage_km_litre', 'Highway_Mileage_km_litre']]
correlation_matrix = mileage_df.corr()
correlation_matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, 
            xticklabels=['City Mileage', 'Highway Mileage'],
            yticklabels=['City Mileage', 'Highway Mileage'])
plt.title('Correlation between City and Highway Mileage')
plt.show()
if correlation_matrix.iloc[0, 1] < 0:
    print("There is a negative correlation between City Mileage and Highway Mileage.")
else:
    print("There is a positive correlation between City Mileage and Highway Mileage.")
```

```python
#TASK14 #Compare the performance of cars based on factors like kerb weight and gross vehicle weight.

#Hint: Calculate and display the mean, median, standard deviation, and coefficient of variation for each manufacturer.
import pandas as pd

# Assuming df is your DataFrame with 'Make', 'Kerb_Weight', and 'Gross_Vehicle_Weight' columns

# Group by 'Make' and select the relevant columns
df_grouped = df.groupby('Make')[['Kerb_Weight', 'Gross_Vehicle_Weight']]

# Calculate the mean, median, std, and cv (Coefficient of Variation)
mean = df_grouped.mean()
median = df_grouped.median()
std = df_grouped.std()
cv = (std / mean) * 100  # Coefficient of Variation
print('Mean:',mean)
print('=============================================')
print('Median:',median)
print('=============================================')
print('Standard deviation:',std)
print('=============================================')
print('Coefficient of variance:',cv)

```
